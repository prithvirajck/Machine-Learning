```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
from scipy import special

py.offline.init_notebook_mode(connected=True)
```

```python
#Read data into a dataframe
df = pd.read_csv("C:/Prithvi/sampleCarbonEmmission.csv", index_col='monthYear') 
```

```python
df.head(12)
```

```python
#Plot Value
data = [go.Scatter(x=df.index, y=df['Value'], name = "Carbon Emmissions in Million Metric Tons")]

#layout = dict(title = "Monthly carbon emissions from electricity generation", xaxis = df.index)
#fig = dict(data=value, layout=layout)
py.offline.iplot(data)
```

```python
#Calculate rolling standard deviation
value_std = pd.rolling_std(df['Value'], 2)
```

```python
value = go.Scatter(x=df.index, y=df['Value'])
std = go.Scatter(x=df.index, y=value_std)

#Plot "Values" column and its standard deviation
data = [value, std]

layout = go.Layout(
    title = 'Carbon Emmissions in Million Metric Tons',
    yaxis = dict(
        title='Carbon Emmissions'
    ),
    xaxis = dict(
        title='Months'
    )
)
fig = go.Figure(data=data, layout=layout)
py.offline.iplot(fig)
```

```python
#Create an anomaly series when the rolling standard deviation is greater than the standard deviation of the whole set
anomaly_data = pd.DataFrame(value_std < df['Value'].std())
```

```python
anomaly_data.columns = ['anomaly']
anomaly_data['anomaly'] = np.where(anomaly_data['anomaly'] == False, 350, 0)

#lead = go.Scatter(x=df.index, y=df['Value'], name = "Carbon emissions")
anomaly = go.Scatter(x=anomaly_data.index, y=anomaly_data['anomaly'], name = "anomaly")

data = [value, anomaly]
layout = go.Layout(
    title = 'Carbon Emmissions in Million Metric Tons',
    yaxis = dict(
        title='Carbon Emmissions'
    ),
    xaxis = dict(
        title='Months'
    )
)
fig = go.Figure(data=data, layout=layout)
py.offline.iplot(fig)
```
